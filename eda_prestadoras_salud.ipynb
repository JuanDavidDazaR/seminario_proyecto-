```python
import msoffcrypto
import io
import pandas as pd
import numpy as np
import os

# Desencriptar el archivo
decrypted_buffer = io.BytesIO()
with open("data\dane\anex-EEVV-DefuncionesNoFetalesCuadroAnoCorrido-2025pr.xlsx", "rb") as f:
    office_file = msoffcrypto.OfficeFile(f)
    office_file.load_key(password="VelvetSweatshop")
    office_file.decrypt(decrypted_buffer)

# Leer todas las hojas del archivo
xls = pd.ExcelFile(decrypted_buffer)
print(f"Hojas encontradas: {xls.sheet_names}")

# Crear carpeta para guardar los datasets si no existe
if not os.path.exists('datasets_defunciones'):
    os.makedirs('datasets_defunciones')
    print("\n📁 Carpeta 'datasets_defunciones' creada")

# Diccionario para guardar todos los dataframes
dataframes = {}

# Función para limpiar nombres de columnas
def limpiar_nombre_columna(nombre):
    if pd.isna(nombre) or nombre == '' or nombre == 'nan':
        return 'columna'
    return str(nombre).strip().replace('\n', ' ').replace('  ', ' ')

# Procesar cada hoja
for sheet_name in xls.sheet_names:
    print(f"\n{'='*60}")
    print(f"PROCESANDO HOJA: {sheet_name}")
    print('='*60)
    
    # Leer la hoja completa
    df_sheet = pd.read_excel(decrypted_buffer, sheet_name=sheet_name, header=None)
    
    # Mostrar estructura
    print(f"\n📊 Dimensiones de la hoja: {df_sheet.shape}")
    print("Primeras 5 filas (vista previa):")
    for i in range(min(5, len(df_sheet))):
        print(f"  Fila {i}: {df_sheet.iloc[i, :3].tolist()}")
    
    # BUSCAR LA ESTRUCTURA DE CADA HOJA
    print("\n🔍 Analizando estructura...")
    
    # Encontrar dónde empiezan los datos reales
    fila_inicio_datos = None
    fila_encabezados = None
    
    for i in range(len(df_sheet)):
        fila = df_sheet.iloc[i].fillna('').astype(str)
        
        # Buscar patrones que indican inicio de datos
        texto_fila = ' '.join(fila[:5].tolist()).upper()
        
        if any(patron in texto_fila for patron in ['TOTAL NACIONAL', 'DEPARTAMENTO', 'GRUPOS DE EDAD', 'TOTAL', 'HOMBRES', 'MUJERES']):
            # Esta fila podría ser encabezado o datos
            if i > 0:
                # Verificar si la fila anterior tiene títulos
                fila_anterior = df_sheet.iloc[i-1].fillna('').astype(str)
                texto_anterior = ' '.join(fila_anterior[:5].tolist())
                
                if 'CUADRO' in texto_anterior.upper():
                    # La fila actual son los encabezados
                    fila_encabezados = i
                    fila_inicio_datos = i + 1
                    print(f"  ✅ Encabezados encontrados en fila {i}")
                    print(f"  ✅ Datos comienzan en fila {i+1}")
                    break
                else:
                    # La fila actual podría ser tanto encabezados como datos
                    fila_encabezados = i
                    fila_inicio_datos = i + 1
                    print(f"  ⚠️ Posibles encabezados en fila {i}")
                    break
    
    # Si no se encontró estructura, buscar por 'Cuadro'
    if fila_inicio_datos is None:
        for i in range(len(df_sheet)):
            fila = df_sheet.iloc[i, :2].astype(str).tolist()
            texto = ' '.join([str(x) for x in fila])
            if "Cuadro" in texto and "Defunciones" in texto:
                # El cuadro comienza aquí, los datos están 3-4 filas después
                fila_inicio_datos = i + 3
                fila_encabezados = i + 2
                print(f"  ✅ Cuadro encontrado en fila {i}")
                print(f"  ✅ Datos desde fila {fila_inicio_datos}")
                break
    
    # Extraer los datos
    if fila_inicio_datos is not None and fila_inicio_datos < len(df_sheet):
        try:
            # Obtener encabezados
            if fila_encabezados is not None and fila_encabezados < len(df_sheet):
                encabezados_raw = df_sheet.iloc[fila_encabezados].fillna('').tolist()
            else:
                encabezados_raw = df_sheet.iloc[fila_inicio_datos - 1].fillna('').tolist()
            
            # Limpiar encabezados
            encabezados = [limpiar_nombre_columna(h) for h in encabezados_raw]
            
            # Crear diccionario para evitar columnas duplicadas
            encabezados_unicos = []
            contadores = {}
            
            for h in encabezados:
                if h in contadores:
                    contadores[h] += 1
                    encabezados_unicos.append(f"{h}_{contadores[h]}")
                else:
                    contadores[h] = 0
                    encabezados_unicos.append(h)
            
            # Extraer datos
            datos = df_sheet.iloc[fila_inicio_datos:].copy()
            datos.columns = encabezados_unicos
            
            # Limpiar datos
            # Eliminar filas totalmente vacías
            datos = datos.dropna(how='all')
            
            # Eliminar columnas totalmente vacías
            datos = datos.dropna(axis=1, how='all')
            
            # Eliminar filas que son solo títulos o notas
            datos = datos[~datos.iloc[:, 0].astype(str).str.contains('Volver al índice|Cuadro|Fuente|Notas', na=False)]
            
            # Resetear índice
            datos = datos.reset_index(drop=True)
            
            # Guardar en el diccionario
            nombre_dataset = f"defunciones_{sheet_name.lower()}"
            dataframes[nombre_dataset] = datos
            
            print(f"\n📦 DATASET CREADO: {nombre_dataset}")
            print(f"   • Filas: {len(datos)}")
            print(f"   • Columnas: {len(datos.columns)}")
            print(f"   • Columnas: {datos.columns.tolist()}")
            
            # Mostrar muestra
            print("\n   Muestra de datos (primeras 3 filas):")
            print(datos.head(3).to_string())
            
            # Guardar a CSV
            archivo_csv = f"datasets_defunciones/{nombre_dataset}.csv"
            datos.to_csv(archivo_csv, index=False, encoding='utf-8-sig')
            print(f"   💾 Guardado en: {archivo_csv}")
            
        except Exception as e:
            print(f"❌ Error al procesar {sheet_name}: {e}")
            print("   Continuando con la siguiente hoja...")
    else:
        print(f"⚠️ No se pudo determinar la estructura de datos para {sheet_name}")

# RESUMEN FINAL
print("\n" + "="*60)
print("RESUMEN FINAL - DATASETS CREADOS")
print("="*60)

total_filas = 0
for nombre, df in dataframes.items():
    print(f"\n📊 {nombre}:")
    print(f"   • Filas: {len(df):,}")
    print(f"   • Columnas: {len(df.columns)}")
    print(f"   • Archivo: datasets_defunciones/{nombre}.csv")
    total_filas += len(df)

print(f"\n✅ TOTAL: {len(dataframes)} datasets creados")
print(f"✅ TOTAL DE FILAS COMBINADAS: {total_filas:,}")
print(f"📁 Los archivos están en la carpeta 'datasets_defunciones/'")

# Mostrar todos los datasets disponibles
print("\n📋 DATASETS DISPONIBLES:")
for i, nombre in enumerate(dataframes.keys(), 1):
    print(f"  {i}. {nombre}")
```

```python
import pandas as pd
import os
from pathlib import Path

carpeta_base = 'datasets_defunciones_completos'

# Función para inspeccionar un cuadro específico de todos los años
def inspeccionar_cuadro(nombre_cuadro):
    """
    Inspecciona un cuadro específico (ej: 'cuadro1', 'cuadro5', etc.) en todos los años
    """
    print(f"\n{'='*70}")
    print(f"INSPECCIONANDO: {nombre_cuadro.upper()} - TODOS LOS AÑOS")
    print('='*70)
    
    for año in ['2022', '2023', '2024', '2025']:
        carpeta_año = os.path.join(carpeta_base, f'año_{año}')
        if not os.path.exists(carpeta_año):
            continue
            
        # Buscar archivos que coincidan con el cuadro
        archivos = [f for f in os.listdir(carpeta_año) 
                   if f.endswith('.csv') and nombre_cuadro.lower() in f.lower()]
        
        for archivo in archivos:
            ruta = os.path.join(carpeta_año, archivo)
            df = pd.read_csv(ruta)
            
            print(f"\n📁 AÑO {año} - {archivo}")
            print(f"   • Filas: {len(df):,}")
            print(f"   • Columnas: {len(df.columns)}")
            print(f"   • Columnas: {df.columns.tolist()[:10]}...")  # Primeras 10 columnas
            print(f"   • Tipos de datos:\n{df.dtypes.value_counts()}")
            print(f"\n   Primeras 3 filas:")
            print(df.head(3).to_string())
            print("-" * 40)

# Inspeccionar Cuadro 1 (defunciones por área y sexo)
inspeccionar_cuadro('cuadro1')

# Inspeccionar Cuadro 5 (defunciones por depto y causas)
inspeccionar_cuadro('cuadro5')

# Inspeccionar Cuadro 11 (el más grande)
inspeccionar_cuadro('cuadro11')
```

```python
import pandas as pd
import os
import glob

carpeta_base = 'datasets_defunciones_completos'

def unir_cuadro_especifico(nombre_cuadro):
    """
    Une un cuadro específico de todos los años
    Ej: 'cuadro1', 'cuadro5', 'cuadro11', etc.
    """
    print(f"\n{'='*70}")
    print(f"UNIENDO {nombre_cuadro.upper()} - TODOS LOS AÑOS")
    print('='*70)
    
    dataframes = []
    total_filas = 0
    
    for año in ['2022', '2023', '2024', '2025']:
        carpeta_año = os.path.join(carpeta_base, f'año_{año}')
        patron = os.path.join(carpeta_año, f'*{nombre_cuadro.lower()}*.csv')
        archivos = glob.glob(patron)
        
        if archivos:
            archivo = archivos[0]  # Tomar el primero
            df = pd.read_csv(archivo)
            
            # Asegurar que la columna 'año' existe
            if 'año' not in df.columns:
                df['año'] = año
            
            dataframes.append(df)
            print(f"  ✅ Año {año}: {len(df):,} filas")
            total_filas += len(df)
        else:
            print(f"  ❌ Año {año}: No encontrado")
    
    if dataframes:
        # Unir todos
        df_unido = pd.concat(dataframes, ignore_index=True)
        
        print(f"\n📊 TOTAL UNIDO:")
        print(f"   • Filas totales: {len(df_unido):,}")
        print(f"   • Columnas: {len(df_unido.columns)}")
        print(f"   • Distribución por año:\n{df_unido['año'].value_counts().sort_index()}")
        
        # Guardar
        nombre_archivo = f"todos_los_{nombre_cuadro}.csv"
        df_unido.to_csv(nombre_archivo, index=False, encoding='utf-8-sig')
        print(f"\n💾 Guardado en: {nombre_archivo}")
        
        return df_unido
    else:
        print("❌ No se encontraron datos")
        return None

# UNIR TODOS LOS CUADROS QUE QUIERAS:

# 1. Unir Cuadro 1
df_cuadro1 = unir_cuadro_especifico('cuadro1')

# 2. Unir Cuadro 5 (el más grande)
df_cuadro5 = unir_cuadro_especifico('cuadro5')

# 3. Unir Cuadro 11
df_cuadro11 = unir_cuadro_especifico('cuadro11')

# 4. Unir Cuadro 3
df_cuadro3 = unir_cuadro_especifico('cuadro3')
```

```python
import pandas as pd
import os
import glob

# ============================================
# CONFIGURACIÓN
# ============================================
carpeta_base = 'data/datos_completos/nacimientos'

def inspeccionar_estructura_nacimientos():
    """
    Inspecciona la estructura de todos los cuadros de nacimientos
    Muestra número de columnas, nombres y una muestra de datos
    """
    print("\n" + "="*80)
    print("📊 INSPECCIÓN DE ESTRUCTURA - DATOS DE NACIMIENTOS")
    print("="*80)
    
    # Lista de cuadros de nacimientos
    cuadros_nacimientos = ['cuadro1', 'cuadro3', 'cuadro5', 'cuadro6a', 'cuadro11', 'cuadro13']
    años = ['2022', '2023', '2024', '2025']
    
    resumen_estructura = []
    
    for cuadro in cuadros_nacimientos:
        print(f"\n{'#'*60}")
        print(f"📋 ANALIZANDO {cuadro.upper()}")
        print('#'*60)
        
        estructuras_por_año = {}
        
        for año in años:
            carpeta_año = os.path.join(carpeta_base, f'año_{año}')
            patron = os.path.join(carpeta_año, f'nacimientos_{año}_{cuadro}*.csv')
            archivos = glob.glob(patron)
            
            if archivos:
                archivo = archivos[0]
                df = pd.read_csv(archivo, nrows=5)  # Leer solo 5 filas para inspección
                
                # Información básica
                print(f"\n📁 AÑO {año}:")
                print(f"   • Archivo: {os.path.basename(archivo)}")
                print(f"   • Filas totales: {len(pd.read_csv(archivo)):,}")
                print(f"   • Columnas: {len(df.columns)}")
                
                # Mostrar primeras columnas
                print(f"   • Primeras 10 columnas: {df.columns[:10].tolist()}")
                
                # Tipos de datos
                print(f"   • Tipos de datos:")
                tipos = df.dtypes.value_counts()
                for tipo, count in tipos.items():
                    print(f"      - {tipo}: {count} columnas")
                
                # Guardar para resumen
                estructuras_por_año[año] = {
                    'columnas': len(df.columns),
                    'nombres_columnas': list(df.columns),
                    'tipos': dict(df.dtypes.value_counts())
                }
                
                # Mostrar muestra de datos
                print(f"\n   📝 Muestra de datos (primeras 3 filas):")
                print(df.head(3).to_string())
                print("-" * 40)
            else:
                print(f"\n❌ AÑO {año}: No encontrado")
        
        # Resumen del cuadro
        if estructuras_por_año:
            print(f"\n📊 RESUMEN {cuadro.upper()} - COMPARATIVA POR AÑOS:")
            
            # Verificar consistencia de columnas entre años
            columnas_por_año = {año: info['columnas'] for año, info in estructuras_por_año.items()}
            print(f"   • Columnas por año: {columnas_por_año}")
            
            # Verificar si todas tienen las mismas columnas
            if len(set(columnas_por_año.values())) == 1:
                print(f"   ✅ TODOS LOS AÑOS TIENEN {list(columnas_por_año.values())[0]} COLUMNAS")
            else:
                print(f"   ⚠️ DIFERENTE NÚMERO DE COLUMNAS ENTRE AÑOS")
            
            # Guardar en resumen general
            resumen_estructura.append({
                'cuadro': cuadro,
                'columnas_2022': estructuras_por_año.get('2022', {}).get('columnas', 0),
                'columnas_2023': estructuras_por_año.get('2023', {}).get('columnas', 0),
                'columnas_2024': estructuras_por_año.get('2024', {}).get('columnas', 0),
                'columnas_2025': estructuras_por_año.get('2025', {}).get('columnas', 0)
            })
    
    # ============================================
    # RESUMEN GENERAL
    # ============================================
    print("\n" + "="*80)
    print("📈 RESUMEN GENERAL - ESTRUCTURA DE NACIMIENTOS")
    print("="*80)
    
    resumen_df = pd.DataFrame(resumen_estructura)
    print(resumen_df.to_string(index=False))
    
    # Guardar resumen
    resumen_df.to_csv('estructura_nacimientos.csv', index=False, encoding='utf-8-sig')
    print(f"\n💾 Resumen guardado en: estructura_nacimientos.csv")
    
    return resumen_df

# ============================================
# FUNCIÓN PARA VER DETALLE DE COLUMNAS
# ============================================
def ver_detalle_columnas_nacimientos(cuadro_especifico=None):
    """
    Muestra el detalle completo de columnas para cada cuadro
    """
    print("\n" + "="*80)
    print("🔍 DETALLE DE COLUMNAS POR CUADRO")
    print("="*80)
    
    cuadros = [cuadro_especifico] if cuadro_especifico else ['cuadro1', 'cuadro3', 'cuadro5', 'cuadro6a', 'cuadro11', 'cuadro13']
    año_referencia = '2024'  # Usar 2024 como referencia
    
    for cuadro in cuadros:
        carpeta_año = os.path.join(carpeta_base, f'año_{año_referencia}')
        patron = os.path.join(carpeta_año, f'nacimientos_{año_referencia}_{cuadro}*.csv')
        archivos = glob.glob(patron)
        
        if archivos:
            archivo = archivos[0]
            df = pd.read_csv(archivo)
            
            print(f"\n{'#'*60}")
            print(f"📋 {cuadro.upper()} - AÑO {año_referencia}")
            print('#'*60)
            print(f"Total columnas: {len(df.columns)}")
            
            # Mostrar todas las columnas en grupos
            print("\n📌 LISTADO COMPLETO DE COLUMNAS:")
            for i, col in enumerate(df.columns, 1):
                print(f"  {i:3d}. {col}")
            
            # Mostrar tipos de datos
            print(f"\n📊 TIPOS DE DATOS:")
            tipos = df.dtypes.value_counts()
            for tipo, count in tipos.items():
                print(f"  • {tipo}: {count} columnas")
            
            # Mostrar estadísticas básicas de las primeras columnas numéricas
            print(f"\n📈 ESTADÍSTICAS BÁSICAS (primeras columnas numéricas):")
            columnas_numericas = df.select_dtypes(include=[np.number]).columns[:5]
            if len(columnas_numericas) > 0:
                print(df[columnas_numericas].describe().to_string())
        else:
            print(f"\n❌ {cuadro} no encontrado para año {año_referencia}")

# ============================================
# FUNCIÓN PARA VERIFICAR CONSISTENCIA ENTRE AÑOS
# ============================================
def verificar_consistencia_nacimientos():
    """
    Verifica si las columnas son consistentes entre años para cada cuadro
    """
    print("\n" + "="*80)
    print("🔄 VERIFICACIÓN DE CONSISTENCIA ENTRE AÑOS")
    print("="*80)
    
    cuadros = ['cuadro1', 'cuadro3', 'cuadro5', 'cuadro6a', 'cuadro11', 'cuadro13']
    años = ['2022', '2023', '2024', '2025']
    
    for cuadro in cuadros:
        print(f"\n📋 {cuadro.upper()}:")
        
        columnas_por_año = {}
        for año in años:
            carpeta_año = os.path.join(carpeta_base, f'año_{año}')
            patron = os.path.join(carpeta_año, f'nacimientos_{año}_{cuadro}*.csv')
            archivos = glob.glob(patron)
            
            if archivos:
                df = pd.read_csv(archivos[0], nrows=1)
                columnas_por_año[año] = set(df.columns)
                print(f"  • {año}: {len(df.columns)} columnas")
        
        # Verificar consistencia
        if len(columnas_por_año) == len(años):
            primera = list(columnas_por_año.values())[0]
            todas_iguales = all(cols == primera for cols in columnas_por_año.values())
            
            if todas_iguales:
                print(f"  ✅ CONSISTENTE: Mismas columnas en todos los años")
            else:
                print(f"  ⚠️ INCONSISTENTE: Diferencias entre años")
                
                # Mostrar diferencias
                for año, cols in columnas_por_año.items():
                    if cols != primera:
                        print(f"     {año} difiere: {primera - cols} (faltan) | {cols - primera} (extra)")

# ============================================
# EJECUTAR INSPECCIÓN
# ============================================
if __name__ == "__main__":
    import numpy as np
    
    print("\n" + "🚀"*20)
    print("   INSPECCIÓN DE DATOS DE NACIMIENTOS")
    print("🚀"*20)
    
    # 1. Inspección general de estructura
    resumen = inspeccionar_estructura_nacimientos()
    
    # 2. Ver detalle de columnas para cuadros específicos
    print("\n" + "🔍"*20)
    print("   DETALLE POR CUADRO")
    print("🔍"*20)
    
    # Ver cada cuadro individualmente
    for cuadro in ['cuadro1', 'cuadro3', 'cuadro5', 'cuadro6a', 'cuadro11', 'cuadro13']:
        ver_detalle_columnas_nacimientos(cuadro)
    
    # 3. Verificar consistencia entre años
    verificar_consistencia_nacimientos()
    
    print("\n" + "✅"*20)
    print("   INSPECCIÓN COMPLETADA")
    print("✅"*20)
```
